Relational Databases store data that is inherently relational.  With the ongoing evolution of the internet and mobility (connected devices), many of the data used in contemporary applications and use-cases don't fit very well into relational schemas.  Additionally, as the need for so-called "big data" increases, the demand for large-scale queries increases.  Because of the inherent volume of data concerned, performance of a SQL database engine would not serve the use-case well.  NoSQL databases address this requirement by leveraging distrubuted computing, so that partitions of the data can be queried in parallel, thereby DRASTICALLY improving reponse time and performance in general.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Each node is responsible for serving/managing a particular partition of the entire data set - i.e. only a subset.  A master server tracks which node is serving up queries for which partition.  This is called the "Map" step (in the MapReduce methodology).  When a query involves more than one paritition, the queried-parition result sets are later combined by a master server into the final, complete result set in the "Reduce" step.  Nodes also work together to back up one another's data (but still are responsible for exposing only a fixed partition of the data set).  So if a node goes offline, one of the other nodes can temporarily serve up the now-downed node's data partition until a replacement can be erected for the offline node.
